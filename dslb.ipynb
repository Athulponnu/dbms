```python
#4.CurveFitting
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def func(x, m, b): return m * x + b

x = np.linspace(0, 10, 100)
y = 3 * x + 2 + np.random.normal(0, 1, x.shape)

popt, _ = curve_fit(func, x, y, p0=[1, 1])
plt.scatter(x, y, s=10, c='red', label="Noisy Data")
plt.plot(x, func(x, *popt), c='blue', label="Fitted Line")
plt.xlabel("X"); plt.ylabel("Y"); plt.title("Curve Fitting Using SciPy")
plt.legend(); plt.show()
```

```python
#5a.LINEAR REGRESSION
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + 2 * x**2 + np.random.randn(100, 1)

X_poly = np.c_[np.ones((100, 1)), x, x**2]
theta = np.linalg.inv(X_poly.T @ X_poly) @ X_poly.T @ y
y_pred = X_poly @ theta

r2 = 1 - np.sum((y - y_pred) ** 2) / np.sum((y - np.mean(y)) ** 2)

print("Coefficients:", theta.ravel())
print("R²:", r2)

x_sort = np.sort(x, axis=0)
y_sort_pred = np.c_[np.ones_like(x_sort), x_sort, x_sort**2] @ theta

plt.scatter(x, y, c='blue', label='Data')
plt.plot(x_sort, y_sort_pred, c='red', lw=2, label='Fit')
plt.xlabel("X"); plt.ylabel("Y")
plt.title("Polynomial Regression (Degree 2)")
plt.legend(); plt.show()
```

```python
#5b.POLYNOMIAL REGRESSION
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
x = 2 * np.random.rand(100, 1)
y = 2 * np.exp(1.5 * x) + np.random.randn(100, 1)  # Ensure y > 0 by design
x = x[y > 0]
y = y[y > 0]
log_y = np.log(y)
x_b = np.c_[np.ones((x.shape[0], 1)), x]  # Add bias term
theta_best = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(log_y)
a = np.exp(theta_best[0])  # Since intercept = ln(a)
b = theta_best[1]

print("Estimated Coefficients: ", theta_best.ravel())

x_sorted = np.sort(x, axis=0)
y_pred = a * np.exp(b * x_sorted)

ss_total = np.sum((y - np.mean(y)) ** 2)
ss_residual = np.sum((y - a * np.exp(b * x)) ** 2)
r_squared = 1 - (ss_residual / ss_total)
print(f"R-Squared value: {r_squared:.4f}")

plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x_sorted, y_pred, color='red', linewidth=2, label='Exponential Fit')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("Exponential Regression using NumPy")
plt.show()
```

```python
#5c.Exponential Regression
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x = 2 * np.random.rand(100, 1)
y = 2 * np.exp(1.5 * x) + np.random.randn(100, 1)
x, y = x[y > 0], y[y > 0]

log_y = np.log(y)
X_b = np.c_[np.ones((x.shape[0], 1)), x]
theta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ log_y
a, b = np.exp(theta[0]), theta[1]

x_sorted = np.sort(x, axis=0)
y_pred = a * np.exp(b * x_sorted)

r2 = 1 - np.sum((y - a * np.exp(b * x))**2) / np.sum((y - np.mean(y))**2)

print("Coefficients:", theta.ravel())
print(f"R²: {r2:.4f}")

plt.scatter(x, y, c='blue', label='Data')
plt.plot(x_sorted, y_pred, c='red', lw=2, label='Exp Fit')
plt.xlabel("X"); plt.ylabel("Y")
plt.title("Exponential Regression")
plt.legend(); plt.show()
```

```python
#6th
import numpy as np
import matplotlib.pyplot as plt
from sklearn import neighbors, datasets
from matplotlib.colors import ListedColormap

iris = datasets.load_iris()
X, y = iris.data[:, :2], iris.target
clf = neighbors.KNeighborsClassifier(n_neighbors=5).fit(X, y)

h = .02
xx, yy = np.meshgrid(np.arange(X[:, 0].min()-1, X[:, 0].max()+1, h), 
                     np.arange(X[:, 1].min()-1, X[:, 1].max()+1, h))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

plt.pcolormesh(xx, yy, Z, cmap=ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF']))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=ListedColormap(['#FF0000', '#00FF00', '#0000FF']), edgecolor='k', s=20)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('k-NN Classification')
plt.show()
```

```python
#7.Finding of Outliers
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = {'Value': [10, 12, 12, 13, 13, 12, 100, 13, 12, 12, 11]}
df = pd.DataFrame(data)
Q1 = df['Value'].quantile(0.25)
Q3 = df['Value'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df['outliers'] = (df['Value'] < lower_bound) | (df['Value'] > upper_bound)
print("Dataset with outliers:")
print(df)
plt.boxplot(df['Value'])
plt.title("Box Plot to Detect Outliers")
plt.ylabel("Values")
plt.grid(True)
plt.show()


```

```python
#9th

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data = np.random.rand(100, 2)
kmeans = KMeans(n_clusters=3, random_state=0).fit(data)
labels, centroids = kmeans.labels_, kmeans.cluster_centers_

for i in range(3):
    plt.scatter(*data[labels == i].T, label=f'Cluster {i+1}')
plt.scatter(*centroids.T, c='black', marker='X', s=100, label='Centroids')

plt.title("K-Means Clustering")
plt.xlabel("Feature 1"); plt.ylabel("Feature 2")
plt.legend(); plt.grid(True); plt.show()
```
